```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_pickle("./df_preprocessed_text.pkl")
df.head()
```

```python
df_unique = df["preprocessed_text"].str.split(expand=True).stack().value_counts()
df_unique.head()
```

```python
df_unique = df_unique.reset_index()
df_unique.head()
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(df["preprocessed_text"])
```

```python
feature_names = vectorizer.get_feature_names_out()
idf_value = vectorizer.idf_
df_tfidf = pd.DataFrame()
df_tfidf["Feature_name"] = feature_names
df_tfidf["idf_values"] = np.round(idf_value,2)
df_tfidf.head()
```

```python
df_unique.rename(columns={"index":"Feature_name"},inplace=True)
df_unique.head()
```

```python
df_combined = pd.merge(df_tfidf,df_unique,how='inner',on='Feature_name')
df_combined.head(10)
```

```python
df_combined = df_combined.sort_values(by="idf_values",ascending=True)
df_combined.head(10)
```
