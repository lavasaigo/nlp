```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('./df_preprocessed_text.pkl')
df.info()
```

```python
total_text = " ".join(df['preprocessed_text'].astype(str))
words_lst = total_text.split()
uniques_words_lst = list(set(words_lst))
print(f"Total unique words:{len(uniques_words_lst)}")
print(f"First 5 unique words are: {uniques_words_lst[:5]}")
```

```python
from collections import Counter

unique_words_frequency_dict = dict(Counter(words_lst))
unique_words_frequency_dict
```

```python
sorted_unique_words = sorted(unique_words_frequency_dict.items(),key=lambda x:x[1],reverse=True)
sorted_unique_words[:5]
```
